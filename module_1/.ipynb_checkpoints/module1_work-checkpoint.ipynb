```python
Series_nationality = pd.DataFrame(df[df['Value'] > df['Value'].mean()]['Nationality'])
Series_nationality.describe()
```

```python
gk_wage_reflex = df[(df['Position'] == 'GK') & (df['GKReflexes'] == df['GKReflexes'].max())]['Wage'].mean()
gk_wage_handling = df[(df['Position'] == 'GK') & (df['GKHandling'] == df['GKHandling'].max())]['Wage'].mean()
gk_wage = gk_wage_reflex/gk_wage_handling
gk_wage
```

```python
shot_aggression = df[df['Aggression'] == df['Aggression'].max()]['ShotPower'].mean()
shot_non_aggression = df[df['Aggression'] == df['Aggression'].min()]['ShotPower'].mean()
shot_power = shot_aggression/shot_non_aggression
shot_power
```
